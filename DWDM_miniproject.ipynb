# Encoding Data

```python
from sklearn.preprocessing import LabelEncoder

q3_encoder = LabelEncoder()
```

```python
q3_encoder.fit(df['q3'])
```

```python
q3_values = q3_encoder.transform(df['q3'])
```

```python
print("Before Encoding:", list(df['q3'][-10:]))
print("After Encoding:", q3_values[-10:])
print("The inverse from the encoding result:", q3_encoder.inverse_transform(q3_values[-10:]))
```

```python
q_encoder = LabelEncoder()
q_encoder.fit(df['q1'])
q1_values = q_encoder.transform(df['q1'])
print("Before Encoding:", list(df['q1'][-10:]))
print("After Encoding:", q1_values[-10:])
print("The inverse from the encoding result:", q_encoder.inverse_transform(q1_values[-10:]))
```

```python
mf = pd.DataFrame()
mf
```

```python

def encode_df(df):
  for i in range(1,44):
    q_encoder.fit(df['q'+ str(i)])
    q_values = q_encoder.transform(df['q'+ str(i)])
    #print("Before Encoding:", list(df['q'+ str(i)][-10:]))
    mf['q'+ str(i)] = q_values
    print("After Encoding:", q_values[-10:])
    #print("The inverse from the encoding result:", q_encoder.inverse_transform(q_values[-10:]))
```

```python
encode_df(df)
```

```python
X = mf.drop('q43', axis=1)
y = mf['q43']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import SGDClassifier
```

```python
clf = SGDClassifier(loss="hinge", penalty="l2", max_iter=5)
clf.fit(X_train, y_train)
SGDClassifier(max_iter=5)
```

```python
np.random.seed(42)
clf.score(X_test, y_test)
```

**----------------------------------------------------------------------------------------------------------------------------------**

```python

from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators=20, random_state=0)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
regressor.score(X_test, y_test)
```

```python
from sklearn import metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
